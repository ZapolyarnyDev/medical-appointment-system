# Тестирование

## Цель

Тесты должны быстро показывать, что основные сценарии живы, а права доступа не пропускают лишнее

## Что покрыто

- запись пациента на свободный слот
- запрет записи на занятый слот
- запрет записи на прошедший слот
- отмена записи
- перенос записи
- завершение приема врачом
- создание расписания
- регистраторская запись
- разграничение ролей

## Как запускать

Windows:

```powershell
.\gradlew.bat test
```

Linux:

```bash
./gradlew test
```

Для части интеграционных тестов нужен Docker, потому что используется Testcontainers
