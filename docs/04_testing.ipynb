# Стратегия тестирования

Этот документ описывает, что именно должно быть доказано тестами

Команды запуска вынесены в [Тестирование](testing.ipynb)

## Цели

- проверить основные пользовательские сценарии
- доказать ограничения ролей
- поймать ошибки расписания и записи
- быстро понять, что окружение живо перед сдачей

## Уровни тестов

### Юнит и сервисные проверки

Проверяют бизнес-правила без полного окружения

- нельзя записаться на занятый слот
- нельзя записаться на прошедший слот
- нельзя создать расписание в прошлом
- перенос записи не должен создавать дубль

### Интеграционные проверки

Проверяют работу с PostgreSQL через Testcontainers

- создание расписания
- запись пациента
- отмена записи
- перенос записи
- завершение приема
- регистраторская запись

### Проверки безопасности

Проверяют, что роль и доменная привязка работают вместе

- аноним не видит внутреннюю часть
- пациент не видит чужие записи
- пациент не отменяет чужую запись
- врач не видит чужие приемы
- регистратор не попадает в админку
- главный врач попадает в админку
- публичный домен не отдает внутренние маршруты

## Smoke-проверка

Smoke-проверка должна проходить по ключевому пути системы:

1. пациент записывается
2. регистратор находит пациента и управляет записью
3. врач видит свой прием и завершает его
4. главный врач создает расписание
5. production-профиль не открывает документацию API
