# Тестирование

Проект использует JUnit 5, Spring Boot Test и Testcontainers.

## Запуск

```powershell
.\gradlew.bat spotlessApply test --no-daemon
```

## Покрытие

- создание записи на свободный будущий слот;
- отказ при отсутствующем пациенте, враче или слоте;
- отказ при занятом или прошлом слоте;
- защита от повторной записи на один слот;
- patient self-service booking без передачи `patientId`;
- просмотр специализаций, врачей и свободных слотов;
- маппинг ролей Keycloak в Spring authorities.
