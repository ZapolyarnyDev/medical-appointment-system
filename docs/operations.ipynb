# Эксплуатация

## Ежедневные проверки

- приложение отвечает на `/api/health`
- PostgreSQL работает
- Keycloak работает
- внутренний вход открывается только из внутреннего адреса
- публичный адрес не отдает внутренние маршруты
- OpenAPI закрыт в production

## Резервное копирование

Копировать нужно:

- базу PostgreSQL
- env-файлы
- realm Keycloak
- Caddyfile или конфигурацию reverse proxy

Пример dump PostgreSQL:

Windows:

```powershell
docker compose exec postgres pg_dump -U medical_appointment medical_appointment > backup.sql
```

Linux:

```bash
docker compose exec postgres pg_dump -U medical_appointment medical_appointment > backup.sql
```

## Обновление

1. Остановить приложение
2. Сделать backup PostgreSQL
3. Обновить образ приложения
4. Проверить `.env`
5. Запустить приложение
6. Проверить `/api/health`
7. Проверить вход пациента и сотрудника
8. Проверить создание записи

## Частые проблемы

### HTTP 401 на OpenAPI

В production это ожидаемо

Для локального режима проверь:

- `PUBLIC_DOCS_ENABLED=true`
- активный профиль не `prod`

### Ошибка redirect URI

Проверь redirect URI в Keycloak

Для compose с proxy нужны адреса:

- `http://public.localhost:8080/*`
- `http://internal.localhost:8080/*`
- `http://auth.localhost:8080/*`

### Пациент вошел, но не видит кабинет

Проверь связь учетной записи с `patient_accounts`

### Врач не видит приемы

Проверь роль `DOCTOR`

Проверь связь учетной записи с карточкой врача в `staff_accounts`

### Регистратор не может записать

Проверь роль `REGISTRAR`

Проверь, что выбран пациент и свободный слот
