# Тестирование

## Быстрая проверка

Windows:

```powershell
.\gradlew.bat test
```

Linux:

```bash
./gradlew test
```

Проверяется:

- валидация
- бизнес-правила
- безопасность
- интеграционные сценарии

## Security-тесты

Windows:

```powershell
.\gradlew.bat test --tests *Security*
```

Linux:

```bash
./gradlew test --tests '*Security*'
```

Покрытие:

- аноним не видит внутренние страницы
- пациент не видит чужие записи
- пациент не отменяет чужую запись
- регистратор не попадает в админку
- врач не видит чужие приемы
- главный врач попадает в админку
- публичный домен не отдает внутренние маршруты

## Интеграционные проверки

Windows:

```powershell
.\gradlew.bat test --tests *IntegrationTest
```

Linux:

```bash
./gradlew test --tests '*IntegrationTest'
```

Для этих тестов нужен Docker

Testcontainers поднимает PostgreSQL

Покрытие:

- запись на свободный слот
- запрет записи на занятый слот
- запрет записи на прошедший слот
- регистраторская запись пациента
- отмена записи
- перенос записи
- завершение приема
- генерация расписания

## Smoke-сценарий вручную

1. Запустить compose с proxy
2. Открыть http://public.localhost:8080
3. Войти как `patient`
4. Создать запись
5. Отменить запись в кабинете
6. Открыть http://internal.localhost:8080
7. Войти как `doctor`
8. Завершить прием
9. Войти как `registrar`
10. Найти пациента
11. Записать пациента к врачу
12. Войти как `chief-doctor`
13. Создать расписание и проверить отчет
