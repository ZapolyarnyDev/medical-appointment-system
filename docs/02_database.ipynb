# База данных

База данных хранится в PostgreSQL. Миграции выполняются Flyway из `src/main/resources/db/migration`

## Таблицы

- `specializations`: справочник специализаций
- `doctors`: врачи и кабинеты
- `patients`: пациенты, дата рождения, телефон, номер полиса
- `schedule_slots`: слоты расписания врача
- `appointments`: записи пациентов на слоты
- `patient_accounts`: связь Keycloak аккаунта и пациента
- `staff_accounts`: связь Keycloak аккаунта и сотрудника

## Инварианты

- Один слот может иметь только одну активную запись
- Запись создается только на будущий свободный слот
- Пациентский account mapping уникален по username и Keycloak subject
- Сотрудник с ролью `DOCTOR` должен быть связан с карточкой врача
