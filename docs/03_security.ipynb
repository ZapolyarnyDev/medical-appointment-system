# Безопасность

Авторизация построена на Spring Security, OAuth2 Login, OAuth2 Resource Server и Keycloak.

## Роли

- `PATIENT`: кабинет пациента, самостоятельная запись
- `REGISTRAR`: регистратура и внутренние списки
- `CHIEF_DOCTOR`: администрирование
- `DOCTOR`: учетная запись врача для будущего кабинета

## Границы доступа

- Публично: `/`, `/booking/**`, `/css/**`, `/api/health`
- Пациент: `/account/**`, `/api/appointments/my/**`
- Регистратор и главный врач: `/internal/**`
- Главный врач: `/admin/**`.

## OpenAPI

OpenAPI доступен только если `PUBLIC_DOCS_ENABLED=true`. В `prod` документация выключена

## Identity mapping

Keycloak отвечает за аутентификацию и роли. Приложение хранит связь Keycloak аккаунта с доменными сущностями в `patient_accounts` и `staff_accounts`.
