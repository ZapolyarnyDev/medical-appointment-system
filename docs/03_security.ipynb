# Модель доступа

Этот документ описывает проектные границы безопасности

Пошаговая настройка SSO вынесена в [Безопасность и SSO](security-sso.ipynb)

## Принцип

Приложение не является центром идентификации

Аутентификацию выполняет Keycloak, а приложение принимает результат через OAuth2 Login и JWT

## Роли

- `PATIENT` — пациент, личный кабинет и самостоятельная запись
- `REGISTRAR` — регистратура и запись пациентов
- `DOCTOR` — рабочее место врача
- `CHIEF_DOCTOR` — управление расписанием и административные функции

## Доменные связи

Роль из Keycloak сама по себе не дает доступ к чужим данным

Связь с предметной областью хранится отдельно:

- `patient_accounts` связывает пользователя Keycloak с пациентом
- `staff_accounts` связывает пользователя Keycloak с сотрудником и врачом

## Границы маршрутов

- публичные страницы: `/`, `/booking/**`, `/account/**`
- внутренние страницы: `/internal/**`
- администрирование: `/admin/**`
- служебная проверка: `/api/health`

Внутренние маршруты дополнительно отсекаются reverse proxy на публичном домене

## Правила защиты данных

- пациент не передает `patientId` при самостоятельной записи
- текущий пациент определяется сервером по учетной записи
- врач видит только свои приемы
- регистратор не получает права главного врача
- OpenAPI выключается в production

## Почему так

Такой подход отделяет вход от бизнес-правил

Больница может менять AD, FreeIPA, LDAP или внешний OIDC без переписывания приложения
