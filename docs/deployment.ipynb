# Запуск и окружение

## Локальный запуск через Gradle

Windows:

```powershell
.\gradlew.bat bootRun
```

Linux:

```bash
./gradlew bootRun
```

Адрес приложения:

- http://localhost:8080

## Compose без reverse proxy

Windows:

```powershell
Copy-Item .env.local.example .env
docker compose up --build
```

Linux:

```bash
cp .env.local.example .env
docker compose up --build
```

Состав окружения:

- приложение
- PostgreSQL
- Keycloak

## Compose с reverse proxy

Windows:

```powershell
Copy-Item .env.local.example .env
docker compose --profile proxy up --build
```

Linux:

```bash
cp .env.local.example .env
docker compose --profile proxy up --build
```

Адреса:

- публичная часть: http://public.localhost:8080
- внутренняя часть: http://internal.localhost:8080
- Keycloak: http://auth.localhost:8080

Proxy-правила:

- `public.localhost` отдает публичные страницы
- `public.localhost/internal/**` возвращает 404
- `public.localhost/admin/**` возвращает 404
- `internal.localhost` отдает внутренние страницы после входа
- `auth.localhost` проксирует Keycloak

## Подготовка production

1. Скопировать `.env.prod.example`
2. Задать секреты
3. Задать публичный и внутренний адрес
4. Настроить HTTPS на reverse proxy
5. Установить `PUBLIC_DOCS_ENABLED=false`
6. Установить `SPRING_PROFILES_ACTIVE=prod`
7. Проверить redirect URI в Keycloak

Windows:

```powershell
Copy-Item .env.prod.example .env
docker compose --profile proxy up --build -d
```

Linux:

```bash
cp .env.prod.example .env
docker compose --profile proxy up --build -d
```

## Healthcheck

Dockerfile и compose используют endpoint здоровья приложения

Windows:

```powershell
Invoke-WebRequest http://localhost:8080/api/health
```

Linux:

```bash
curl http://localhost:8080/api/health
```

## Остановка

Windows и Linux:

```bash
docker compose down
```

С удалением volumes:

Windows и Linux:

```bash
docker compose down -v
```
