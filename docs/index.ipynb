# Документация проекта

Этот каталог содержит проектную документацию по системе записи на прием.

## Разделы

- `01_domain_analysis.ipynb` - предметная область и сценарии.
- `02_database.ipynb` - модель данных и миграции.
- `03_security.ipynb` - OAuth2/OIDC, Keycloak, роли и границы доступа.
- `04_testing.ipynb` - тестовая стратегия.
- `05_quality.ipynb` - качество, форматирование и эксплуатационные требования.
- `diagrams/` - диаграммы проекта.
