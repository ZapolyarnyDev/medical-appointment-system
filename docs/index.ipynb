## Быстрый маршрут проверки

1. Запустить окружение через compose
2. Открыть публичный адрес и записать пациента
3. Открыть внутренний адрес и проверить регистратуру
4. Войти врачом и завершить прием
5. Войти главным врачом и проверить расписание
6. Запустить тесты
