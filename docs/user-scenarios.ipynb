# Сценарии пользователей

## Пациент

Цель: самостоятельно выбрать врача и записаться на свободное время

Сценарий:

1. Пациент открывает публичную часть
2. Входит через публичный клиент Keycloak
3. Заполняет или проверяет профиль
4. Выбирает специализацию и врача
5. Выбирает свободный слот
6. Подтверждает запись
7. Видит страницу успешной записи
8. При необходимости отменяет запись в личном кабинете

Защита:

- пациент видит только свои записи
- пациент не может отменить чужую запись
- повторная отправка формы не создает дубль
- занятый или прошедший слот возвращает понятную ошибку

## Регистратор

Цель: быстро записывать пациентов и управлять существующими записями

Сценарий:

1. Регистратор открывает внутренний адрес
2. Входит через внутренний клиент Keycloak
3. Ищет пациента по ФИО, телефону или идентификатору
4. Быстро выбирает найденного пациента
5. Подбирает врача и свободное время
6. Создает запись
7. Отменяет или переносит запись при обращении пациента

Защита:

- регистратор не попадает в административные разделы главного врача
- дубли записи на один слот блокируются
- ошибки показываются рядом с действием

## Врач

Цель: видеть свои приемы и отмечать завершение

Сценарий:

1. Врач открывает внутренний адрес
2. Входит через внутренний клиент Keycloak
3. Видит фильтры “сегодня”, “будущие”, “завершенные”
4. Открывает список своих приемов
5. Завершает прием

Защита:

- врач не видит приемы других врачей
- если учетная запись не привязана к карточке врача, показывается понятное состояние

## Главный врач

Цель: управлять расписанием и рабочими параметрами клиники

Сценарий:

1. Главный врач открывает административный раздел
2. Генерирует рабочий день врача
3. Видит сколько слотов создано и сколько пропущено
4. Использует значения по умолчанию из настроек организации
5. Проверяет расписание врача

Защита:

- административный раздел доступен только роли `CHIEF_DOCTOR`
- даты в прошлом не принимаются
