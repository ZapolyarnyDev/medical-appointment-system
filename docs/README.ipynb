# Обзор документации

## Как устроена папка

В `docs` есть два типа документов

- практические инструкции: запуск, настройка, SSO, тестирование, эксплуатация
- проектные материалы `01..05`: предметная область, данные, доступ, тестовая стратегия, качество

Так файлы не дублируют друг друга: инструкции отвечают “как пользоваться и запускать”, проектные материалы отвечают “почему система устроена именно так”
