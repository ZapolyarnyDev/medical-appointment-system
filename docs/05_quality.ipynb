# Правила качества

Этот документ фиксирует правила проекта

Команды форматирования и проверки вынесены в [Тестирование](testing.ipynb) и [README проекта](../README.md)

## Код

- доменные области разделяются по смыслу, а не по техническим слоям
- контроллеры, сервисы и репозитории одной области лежат рядом
- общая инфраструктура живет в `shared`
- временные решения не должны попадать в публичный API
- пользовательские ошибки должны быть понятными

## Конфигурация

- зависимости Gradle ведутся через version catalog
- секреты не хранятся в репозитории
- production-профиль закрывает OpenAPI
- публичный и внутренний адрес задаются настройками
- параметры клиники задаются без правки кода

## Документация

- все тексты хранятся в UTF-8
- команды указываются отдельно для Windows и Linux, если синтаксис отличается
- практические инструкции не дублируют проектные документы
- точки в конце списков используются минимально
- русские человекочитаемые тексты не заменяются англицизмами без необходимости

## UI и UX

- пациентский путь должен завершаться подтверждением и страницей успеха
- внутренние сценарии должны показывать пустые состояния и понятные ошибки
- регистраторский сценарий оптимизирован под скорость
- врач видит только нужные ему приемы
- главный врач работает с расписанием пачками, а не по одному слоту

## Проверка перед коммитом

- форматирование прошло
- тесты проходят
- документация не содержит битых символов
- изменения соответствуют одному смысловому коммиту
