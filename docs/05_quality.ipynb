# Качество

## Форматирование

В проекте используется Spotless

Проверка:

Windows:

```powershell
.\gradlew.bat spotlessCheck
```

Linux:

```bash
./gradlew spotlessCheck
```

Автоисправление:

Windows:

```powershell
.\gradlew.bat spotlessApply
```

Linux:

```bash
./gradlew spotlessApply
```

После установки hook команда `spotlessApply` запускается перед коммитом

## Правила

- тексты и документация хранятся в UTF-8
- настройки Gradle вынесены в version catalog
- доменные области разделены по смыслу, а не по типам классов
- публичные и внутренние маршруты разделены в Spring Security и reverse proxy
- production-профиль не открывает документацию API
