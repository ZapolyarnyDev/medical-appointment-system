# Качество

## Форматирование

Spotless форматирует Java, Gradle, YAML, Markdown и properties-файлы.

```powershell
.\gradlew.bat spotlessApply
```

## Git hooks

```powershell
.\gradlew.bat installGitHooks
```

После установки pre-commit запускает `spotlessApply`.

## Кодовые правила

- Доменная логика распределяется по предметным областям, а не по техническим слоям
- REST ошибки возвращаются в едином формате
- Пациентские сценарии не доверяют идентификаторам пациента из клиен
- Миграции БД выполняются через Flyway
- SQL-доступ реализуется через jOOQ
