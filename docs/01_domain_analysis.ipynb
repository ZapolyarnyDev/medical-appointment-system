# Анализ предметной области

Система поддерживает запись пациента на прием к врачу. Пользователи разделены на публичных пациентов и сотрудников медицинской организации

## Акторы

- Пациент: создает профиль, просматривает врачей, выбирает слот, подтверждает запись, видит свои записи
- Регистратор: создает пациента и записывает его на прием
- Врач: получает основу для будущего личного расписания
- Главный врач: управляет специализациями, врачами, расписанием, статусами записей и identity mapping

## Основные сущности

- `specializations` - медицинские направления
- `doctors` - врачи
- `patients` - пациенты
- `schedule_slots` - временные слоты расписания
- `appointments` - записи на прием
- `patient_accounts` - привязка Keycloak пользователя к пациенту
- `staff_accounts` - привязка Keycloak пользователя к сотруднику

## Ключевое правило безопасности

Пациентская запись не принимает `patientId` с клиента. Сервер автоматически определяет пациента по учётной записи текущего пользователя
