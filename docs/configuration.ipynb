# Настройка приложения

## Основные env-переменные

Файлы-примеры:

- `.env.example`
- `.env.local.example`
- `.env.prod.example`

Ключевые значения:

- `SPRING_PROFILES_ACTIVE`
- `POSTGRES_DB`
- `POSTGRES_USER`
- `POSTGRES_PASSWORD`
- `KEYCLOAK_ADMIN`
- `KEYCLOAK_ADMIN_PASSWORD`
- `APP_PUBLIC_URL`
- `APP_INTERNAL_URL`
- `PUBLIC_DOCS_ENABLED`
- `IDENTITY_MODE`

## Настройки организации

Настройки клиники позволяют не менять код под конкретное внедрение

- название организации
- публичный адрес
- внутренний адрес
- включение документации
- длительность приема по умолчанию
- начало рабочего дня по умолчанию
- конец рабочего дня по умолчанию

Эти значения используются в UI и при генерации расписания

## Режимы входа

`app.identity.mode` поддерживает режимы:

- `local` — локальные пользователи Keycloak
- `ad` — Active Directory
- `freeipa` — FreeIPA
- `ldap` — общий LDAP-каталог
- `external-oidc` — внешний OIDC-провайдер

Приложение не работает напрямую с AD, FreeIPA или LDAP

Поток всегда такой:

1. Пользователь входит через инфраструктуру организации
2. Keycloak получает личность и роли
3. Приложение принимает OIDC/JWT
4. Связь с доменными сущностями идет через `patient_accounts` и `staff_accounts`

## Публичный и внутренний адрес

Публичная часть:

- `/`
- `/booking/**`
- `/account/**`

Внутренняя часть:

- `/internal/**`
- `/admin/**`

Reverse proxy не должен отдавать внутренние маршруты на публичном домене

## Документация API

В локальном режиме документация может быть включена

В production:

```properties
PUBLIC_DOCS_ENABLED=false
```

Swagger UI и `/v3/api-docs` должны быть недоступны
